```
#Importing dataset
df=pd.read_csv('50_Startups.csv')
```

```
df
```

```
# Dividing dataset into dependent and independent matrices
X= df.iloc[:, 0:4].values
```

```
Y=df.iloc[:,4].values
```

```
#Dealing with categorical data
from sklearn.preprocessing import LabelEncoder
```

```
lbl=LabelEncoder()
```

```
X[:,3]=lbl.fit_transform(X[:,3])
```

```
#Dealing with dummy variables
from sklearn.preprocessing import OneHotEncoder
```

```
onehotencoder = OneHotEncoder(categorical_features=[3])
```

```
X=onehotencoder.fit_transform(X).toarray()
```

```
X
```

```
Y
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size=0.2, random_state=0)
```

```
X_train
```

```
X_test
```

```
#Creating and fitting Multiple linear regression model
from sklearn.linear_model import LinearRegression
```

```
regressor=LinearRegression()
```

```
regressor.fit(X_train, Y_train)
```

```
#Predicting
Y_pred=regressor.predict(X_test)
```

```
#Removing least significant variables using backward elimination
import statsmodels.formula.api as sfa 
```
