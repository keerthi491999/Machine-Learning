```
#Dividing dataset into dependent and independent matrices
X= df.iloc[:, 1:2].values
Y= df.iloc[:, 2].values
#1:2 is used just to indicate that X iks not a vector but a matrix
```

```
#Splitting dataset into training and test sets is not done int his problem because, the information we have is very small and the model need to
#highly accurate and it needs to be traikned on all the levels
```

```
#Fitting Linear regression model to the dataset
from sklearn.linear_model import LinearRegression
rg1= LinearRegression()
rg1.fit(X, Y)
```

```
#Fitting plolynomial regression to dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly =poly_reg.fit_transform(X)
rg2 = LinearRegression()
rg2.fit(X_poly, Y)
```

```
import matplotlib.pyplot as plt
```

```
#Results of LinearRegression

plt.scatter(X, Y, color='blue')
plt.plot(X, rg1.predict(X), color='red')
```

```
#Results of Polynomail Regression
plt.scatter(X, Y, color='blue')
plt.plot(X, rg2.predict(X_poly), color='red')
```

```
#Predicted value in Linear regression
rg1.predict([[6.5]])
```

```
#Predicted value in Polynomial regression
rg2.predict(poly_reg.fit_transform([[6.5]]))
```
